```python
# solution for the course
def monthly_interest_rate(annualInterestRate):
    '''
    Calculates Monthly interest rate
    '''
    return annualInterestRate/12.0

def minimum_monthly_payment(balance):
    '''
    Calculates Minimum monthly payment
    '''
    return monthlyPaymentRate * balance

def monthly_unpaid_balance(balance):
    '''
    Calculates Monthly unpaid balance
    '''
    return balance - minimum_monthly_payment(balance)

def updated_balance_each_month(balance):
    '''
    Calculates Updated balance each month
    '''
    return monthly_unpaid_balance(balance) \
        + monthly_interest_rate(annualInterestRate) * monthly_unpaid_balance(balance)



def balance_after_year(balance):
    '''
    Calculate the credit card balance after one year 
    if a person only pays the minimum monthly payment 
    required by the credit card company each month
    '''
    for month in range(12):
        #print(" Month ",month+1," Remaining balance: ", round(updated_balance_each_month(balance),2))
        balance = updated_balance_each_month(balance)
    return balance   

print("Remaining balance: ",round(balance_after_year(balance),2)) 
```
